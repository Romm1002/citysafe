```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Charger le CSV
df = pd.read_csv("history.csv", sep=";")

# 2. Tracer loss et val_loss
plt.figure()
plt.plot(df["loss"],    label="train loss")
plt.plot(df["val_loss"], label="val loss")
plt.xlabel("Époque")
plt.ylabel("Loss")
plt.legend()
plt.title("Courbe de loss")
plt.show()

# 3. Tracer MAE et val_MAE
plt.figure()
plt.plot(df["MAE"],     label="train MAE")
plt.plot(df["val_MAE"], label="val MAE")
plt.xlabel("Époque")
plt.ylabel("MAE")
plt.legend()
plt.title("Courbe de MAE")
plt.show()

```

```python
plt.figure()
plt.semilogy(df["loss"], label="train MSE")
plt.semilogy(df["val_loss"], label="val MSE")
plt.legend(); plt.xlabel("Époque"); plt.ylabel("log(MSE)")
plt.title("Courbe MSE – échelle log")
```
